# **Parte 1: Fundamentos de bases de datos**


# 1.   Fundamentos de bases de datos y para ciencia de datos. 

Las bases de datos son un sistema que permite recopilar información y estructurar datos, guardados en la memoria de una computadora o en la nube al que se puede acceder de varias maneras. Normalmente es controlada por un sistema de gestión de bases de datos, existen diferentes tipos de bases de datos, las más utilizadas comúnmente son las relacionales y no relacionales.

  Hoy en día, la ciencia de datos tiene una importancia de alto impacto para las empresas, permitiendo el incremento de acciones para la mejora constante en la calidad de sus servicios, a través de los datos, ya sea como recopilar, analizar, limpiar y visualizar. Actualmente la industria necesita de los datos, mismos que pertenecen a la vida cotidiana del ser humano, desde el almacenamiento de información, música, audio y demás.

Independiente de todo lo anterior, los datos sin procesar no generar valor alguno, por lo que los científicos de datos se encargan de dar solución a esta problemática mediante herramientas tecnológicas que proporcionan la ayuda idónea que muchas empresas no sabían que necesitaban, permitiendo optimizar la toma de decisiones y la forma en la que se comunican con sus clientes.

La ciencia de datos tiene un futuro por delante gracias a los avances de inteligencia artificial y al aprendizaje automático que han hecho los procesos de datos más eficientes.

# **Parte 2: Selección y limpieza de los Datos en Python**

Primero importamos librerías

```python
import pandas as pd
import numpy as np
```

Cargamos los datos en una variable y los mostramos en el dataframe

```python
data = ('https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv')
df = pd.read_csv(data)
df
```

* Asignamos nombres a las columnas del dataframe, mediante un arreglo.
   * 'ID': 'ID' = ID del cliente

 * 'Amount': 'X1' = Monto de la transacción

 * 'Gender': 'X2' = Género del cliente (1 = masculino; 2 = femenino)

 * 'Education': 'X3' = Educación del cliente (1 = posgrado; 2 = universidad; 3 = secundaria; 4 = otros)

 * 'Marital': 'X4' = Estado civil del cliente (1 = casado; 2 = soltero; 3 = otros)
 'Age': 'X5' = Edad del cliente (en años)

 * 'His1 - His6': 'X6 - X11' = Historial de pagos de los clientes (de abril a septiembre de 2005)

 * 'Bill1 - Bill6': 'X12 - X17' = Monto de la factura de los clientes (de abril a septiembre de 2005)

 * 'Pay1 - Pay6': 'X18 - X23' = Monto de pago de los clientes (de abril a septiembre de 2005)

 * 'Default': 'Y' = Predicción de incumplimiento de pago (0 = no; 1 = sí)

```python
df.columns = ['ID','Amount','Gender','Education','Marital','Age','His1','His2','His3','His4','His5','His6','Bill1','Bill2','Bill3','Bill4','Bill5','Bill6','Pay1','Pay2','Pay3','Pay4','Pay5','Pay6','Default']
df.head()
```

Comprobando los valores nulos en el dataframe, si es True hay valores nulos y si es False no hay valores nulos 

```python
df.isnull().values.any()
```

Visualizamos los valores nulos realizando una suma y ordenando de mayor a menor

```python
df.isnull().sum().sort_values(ascending=False)
```

Eliminamos los valores nulos


```python
df.dropna(inplace=True)
```

```python
df.isna().values.any()
df
```

Revizamos los valores unicos de cada columna

```python
df.nunique()
```

Imprimir la cantidad de valores nulos por columna

```python
df.isnull().sum().sum()
```

Eliminamos la columna ID ya que no es necesaria para el analisis de datos.

En este caso se hizo por buena practica de eliminar columnas que no agreguen valor, para que sea mas facil de procesar. En nuestro caso la base de datos es pequeña pero al tener una muy grande, se eliminan para que sea mas facil de soportar. 


```python
df.drop(['ID'], axis=1, inplace=True)
df
```

Eliminar los valores duplicados en el dataframe 

```python
df.drop_duplicates(inplace=True)
df
```

Imprimir la cantidad de valores nulos por columna

```python
df.isnull().sum()
```

En esta parte usamos una funcion la cual nos da los valores estadisticos del data frame. 


```python
df.describe().T
```

Dentro la información de la columna Education, se observa que hay 14 registros con valor 0, los cuales no corresponden a ninguna de las categorías, por lo que se procede a eliminarlos.

```python
# Education (1 = graduate school; 2 = university; 3 = high school; 4 = others). 

df['Education'].value_counts()
```

Eliminar los valores que no son 1, 2, 3, 4 en la columna Education 

```python
df = df[df['Education'].isin([1,2,3,4])]
df['Education'].value_counts()
```

Dentro la información de la columna Marital, se observa que hay 54 registros con valor 0, los cuales no corresponden a ninguna de las categorías, por lo que se procede a eliminarlos.

```python
# Marital status (1 = married; 2 = single; 3 = others).
df['Marital'].value_counts()
```

Eliminar los valores que no son 1, 2, 3 en la columna Marital

```python
df = df[df['Marital'].isin([1,2,3])]
df['Marital'].value_counts()
```

En esta parte hacemos una descripcion de los valores generales estadisticos de nuestro data frame

```python
df.describe().T
```

En las preguntas es necesario explorar cuales son los datos mas valiosos agregaremos la funcion de matriz de correlación.

```python
df.corr()
#print(df)
```
